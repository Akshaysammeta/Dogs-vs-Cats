```python
train = []

filenames=os.listdir('train\\cats')
labels = []
#print(filenames)
train_dir='train\\cats'
```

```python
for i in filenames[:100]:
    im=cv2.imread(os.path.join(train_dir,i))
    imr=cv2.resize(im,(128,128))
    train.append(imr)
    labels.append(i)
```

```python
plt.imshow(train[-1])
```

```python
print(labels[-1])
```

```python
npt=np.array(train)
npl=np.array(labels)
```

```python
train=torch.tensor(npt)
labels=torch.tensor(npl)
```

```python
dataset = torch.utils.data.TensorDataset(train, labels)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=len(train), shuffle=True)
```
